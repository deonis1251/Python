```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('sample_data/california_housing_test.csv')
```

```python
sns.scatterplot(data = df, x = 'households', y = 'population')
```

```python
sns.relplot(y = 'median_house_value', x = 'longitude', kind = 'line', data = df)
```

```python
sns.histplot(data = df, x = 'housing_median_age')
```

```python
sns.histplot(data = df, x = 'median_house_value', hue = 'housing_median_age') 
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
penguins.columns
```

```python
sns.scatterplot(data = penguins, x = 'bill_length_mm', y = 'bill_depth_mm', hue = 'sex')
```

```python
cols = ['species', 'island', 'bill_length_mm', 'bill_depth_mm',
       'flipper_length_mm', 'body_mass_g', 'sex']
g = sns.PairGrid(penguins[cols])
g.map(sns.scatterplot)
```

```python

penguins.loc[(penguins['bill_length_mm'] > 35)&(penguins['bill_length_mm'] <= 42), 'bill'] = 'middle'
penguins.loc[penguins['bill_length_mm'] > 42, 'bill'] = 'high'
penguins.loc[penguins['bill_length_mm'] < 35, 'bill'] = 'low'
sns.histplot(data = penguins, x = 'bill_length_mm', hue = 'bill')
```

```python
sns.histplot(data = penguins, x = 'flipper_length_mm', hue = 'bill')
```
